```python
#coding: utf-8
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
%matplotlib inline
```

```python
(x_train, t_train), (x_test, t_test) = cifar10.load_data()
x_train = x_train[:100]
t_train = t_train[:100]
```

```python
label_names = ('airplane', 'mobile', 'bird', 'cat', 'deer',
          'dog', 'frog', 'horse','ship', 'truck')

nclasses = 10
for i in range(nclasses):
    targets = np.where(t_train == i)[0]
    plt.title(label_names[i])
    plt.imshow(x_train[targets[0]])
    plt.show()
```
